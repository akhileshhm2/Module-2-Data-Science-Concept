```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
data={'Hours':[1,2,3,4,5,6,7,8,9],
     'Score':[20,25,35,40,50,60,65,75,85]}
df=pd.DataFrame(data)
X=df[['Hours']]
y=df['Score']
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,random_state=42
)
model=LinearRegression()
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print(y_pred)
mae=mean_absolute_error(y_test,y_pred)
print('mae:',mae)
mse=mean_squared_error(y_test,y_pred)
print('mse:',mse)
rmse=np.sqrt(mse)
print("rmse:",rmse)
r2=r2_score(y_test,y_pred)
print('r2:',r2)
```

```python
#evaluation
print("intercept(c)",model.intercept_)
print("Slope(m)",model.coef_)
```

```python
#plot
plt.scatter(X,y,color='blue')
plt.plot(X,model.predict(X),color='red')
plt.xlabel('Hours studied')
plt.ylabel('Score')
plt.title('Linear Regression')
plt.show()
```

```python
#predict just one value
new_hour=7
predicted_score=model.predict([[new_hour]])[0]
print(f'predicted score for {new_hour} Hours={predicted_score:.2f}')
```

```python
#multiple linear regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data = {
    'Size': [2104, 1600, 2400, 1416, 3000, 1985, 1534, 1427],
    'Bedrooms': [3, 3, 3, 2, 4, 4, 3, 3],
    'Age': [15, 10, 20, 18, 30, 8, 15, 20],
    'Price': [400, 330, 369, 232, 540, 350, 315, 178]
}
df=pd.DataFrame(data)
X=df[['Size','Bedrooms','Age']]
y=df['Price']
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,random_state=42
)
model=LinearRegression()
model.fit(X,y)
y_pred=model.predict(X_test)
y_pred
```

```python
plt.scatter(X,y,color='blue')
plt.plot(X,model.predict(X),color='red')
plt.xlabel('size,room,age')
plt.ylabel('price')
plt.title('Linear Regression')
plt.show()
```
