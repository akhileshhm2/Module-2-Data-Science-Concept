```python
#single integration of sin x
import numpy as np
from scipy import integrate
result,error=integrate.quad(np.sin,0,np.pi)
print("integration",result)
print(" estimation error",error)
```

```python
# double integration of f(x,y) dx dy
result,error=integrate.dblquad(lambda y,x:x*y,0,1, lambda x:0, lambda x:2)
print("double integration",result)
```

```python
# t-test single integration
from scipy import stats
sample=[2.3,1.9,2.5,2.8,2.4]
t_stat,p_val=stats.ttest_1samp(sample,0)
print("t_statistic:",t_stat,"p_value:",p_val)
```

```python
#2-sample ttest
from scipy import stats
group1=[20,22,23,25,30]
group2=[28,29,30,32,35]
t_stat,p_val=stats.ttest_ind(group1,group2)
print("t_statistics",t_stat)
print("p_value",p_val)
```

```python
#z-test
import numpy as np
from statsmodels.stats.weightstats import ztest
scores=np.array([52,49,50,51,53,47,48,54,50,52,49,51])
z_stat,p_val=ztest(scores,value=50)
print("mean of scores:",np.mean(scores))
print("z-statistic",z_stat)
print("p_value",p_val)
```

```python
# 2-sample
male_salary=np.array([55,60,52,61,55,62])
female_salary=np.array([50,43,65,68,58])
z_stat,p_val=ztest(male_salary,female_salary)
print("male mean:",np.mean(male_salary))
print("female:",np.mean(female_salary))
print("z-statistics",z_stat)
print("p_value",p_val)
```

```python
#Anova
import numpy as np
from scipy import stats
group1=[12,14,15,10,13]
group2=[22,24,19,23,25]
group3=[32,34,30,29,31]
f_stat,p_val=stats.f_oneway(group1,group2,group3)
print("F-statistics",f_stat)
print("p-value",p_val)
```
