```python
#simple label encoder
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data=pd.DataFrame({
    "Fruits":["Apple","Banana","Mango","Apple","Banana"]
})
le=LabelEncoder()
data["Fruit_encoder"]=le.fit_transform(data["Fruits"])
print(data)
```

```python
#label encoder using multiple columns
data=pd.DataFrame({
    "city":["delhi","mumbai","chennai","delhi","chennai"],
    "gender":["male",'female','female','male','female']
})
le_city=LabelEncoder()
le_gender=LabelEncoder()

data['city_encoder']=le_city.fit_transform(data['city'])
data['gender_encoder']=le_gender.fit_transform(data['gender'])
print(data)
```

```python
#ordinal encoder
from sklearn.preprocessing import OrdinalEncoder
data=pd.DataFrame({
    "size":["small",'medium','large','medium','small']
})
encoder=OrdinalEncoder(categories=[['small','medium','large']])
data['size_encoded']=encoder.fit_transform(data[['size']])
print(data)
```

```python
#one hot using get dummies
data=pd.DataFrame({"color":['red','blue','green','red']})
ohe=pd.get_dummies(data,columns=['color'])
print(ohe)
```

```python
#one hot using another method
import numpy as np 
from sklearn.preprocessing import OneHotEncoder
data=np.array([['red'],['blue'],['green'],['red']])
encoder=OneHotEncoder(sparse_output=False)
ohe=encoder.fit_transform(data)
columns=encoder.get_feature_names_out(['color'])
df_ohe=pd.DataFrame(ohe,columns=columns)
print(df_ohe)
```

```python
#scalar using minmax
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
data={
    'ID':[101,102,103,104,105,106],
    'Name':['A','B','C','D','E','F'],
    'Age':[10,22,25,30,40,50],
    'salary':[15000,20000,30000,40000,60000,80000],
    'expirence':[1,2,3,5,7,10]
}
df=pd.DataFrame(data)
print("Original Dataset:\n")
print(df)
```

```python
cols_to_scale=['Age','salary','expirence']
scaler=MinMaxScaler()
df[cols_to_scale]=scaler.fit_transform(df[cols_to_scale])
print("\nScaled dataset:\n")
print(df)
```

```python
#standardisation 
import pandas as pd
from sklearn.preprocessing import StandardScaler
data={
    'ID':[101,102,103,104,105,106],
    'Name':['A','B','C','D','E','F'],
    'Age':[10,22,25,30,40,50],
    'salary':[15000,20000,30000,40000,60000,80000],
    'expirence':[1,2,3,5,7,10]
}
df=pd.DataFrame(data)
print("Original Dataset:\n")
print(df)
cols_to_standardize=['Age','salary','expirence']
scaler=StandardScaler()
df[cols_to_standardize]=scaler.fit_transform(df[cols_to_standardize])
print('\nStandardized Dataset:\n')
print(df)
```
