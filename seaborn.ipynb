```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.get_dataset_names())
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("total bill vs tip")
plt.show()
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
sns.lineplot(x="year",y="passengers",data=flights)
plt.title("passengers on years")
plt.show
```

```python
#barplot

sns.barplot(x='day',y='tip',data=tips)
plt.show()
```

```python
exercise=sns.load_dataset("exercise")
exercise
```

```python
sns.scatterplot(x='kind',y='pulse',data=exercise)
plt.show()
```

```python
sns.barplot(x='time',y='pulse',data=exercise)
plt.show()
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.scatterplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
sns.barplot(x='species',y='petal_length',data=iris)
plt.show()
```

```python
sns.lineplot(x='species',y='sepal_length',data=iris)
plt.show()
```

```python
sns.countplot(x='day',data=tips)
plt.show()
```

```python
#countplot
sns.countplot(x='day',data=tips)
plt.show()
```

```python
sns.countplot(x='day',hue='sex',data=tips)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
sns.histplot(x='total_bill',data=tips,kde=True)
plt.show()
```

```python
#kdeplot
sns.kdeplot(x='total_bill',data=tips)
plt.show()
```

```python
#box plot
sns.boxplot(x='day',y='tip',data=tips)
plt.show()
```

```python
#violin plot
sns.violinplot(x='day',y='tip',data=tips)
plt.show()
```

```python
#heatmap
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
sns.pairplot(tips)
```

```python
flights=sns.load_dataset('flights')
corr=flights.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
