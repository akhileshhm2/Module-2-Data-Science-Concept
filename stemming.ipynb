```python
#porterstemmer
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
words=['playing','studies','caring','happiness']
stems=[stemmer.stem(word)for word in words]
print(stems)
```

```python
#lancaster
from nltk import LancasterStemmer
stemmer=LancasterStemmer()
print(stemmer.stem('studies'))
print(stemmer.stem('playing'))
```

```python
#snowball
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
print(stemmer.stem('studies'))
print(stemmer.stem('playing'))
```

```python
#texrblob
from textblob import Word
words=['running','studies','carring','happiness']
for w in words:
    word=Word(w)
    print(f"{w:10}->{word.stem()}")
```

```python
#lemmetizer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
Lemmatizer=WordNetLemmatizer()
words=['running','studies','caring','better','happier']
for w in words:
    print(f'{w:10}->{Lemmatizer.lemmatize(w)}')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
import nltk
Lemmatizer=WordNetLemmatizer()
words=['running','studies','caring','better','happier']
for w in words:
    print(f"{w:10}->{Lemmatizer.lemmatize(w,pos='v')}")
```
